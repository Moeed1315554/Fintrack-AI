```python
cur.execute('''
        CREATE TABLE INCOMEPROFILE(
            PROFILE_ID INT PRIMARY KEY,
            USER_ID INT,
            INCOME_TYPE VARCHAR(40) CHECK(INCOME_TYPE IN("SALARIED","PROFESSIONAL","BUSINESS","OTHERS")),
            MONTHLY_INCOME FLOAT,
            ADDITIONAL_INCOME_TYPE VARCHAR(50) CHECK(ADDITIONAL_INCOME_TYPE IN("STOCK","INVESTEMENTS","BUSINESS","OTHERS")),
            ADDITIONAL_MONTHLY_INCOME FLOAT,
            DEPENDANTS INT CHECK(DEPENDANTS<20),
            CREATED_AT DATETIME,
            UPDATED_AT DATETIME,
            FOREIGN KEY(USER_ID) REFERENCES USER(USER_ID)
        )
        ''')
```

```python
cur.execute('''
        CREATE TABLE EXPENSEPROFILE(
            EXPENSE_ID INT PRIMARY KEY,
            USER_ID INT,
            GROCERIES FLOAT,
            TRAVEL FLOAT,
            MEDFIT FLOAT,
            LEP FLOAT,
            MONTHLY_RENT FLOAT,
            M_BILLS FLOAT,
            FASHION FLOAT,
            ENTERTAINMENT FLOAT,
            EDUCATION FLOAT,
            EMSAVING FLOAT,
            MISCELLANEOUS FLOAT,
            CREATED_AT DATETIME,
            FOREIGN KEY(USER_ID) REFERENCES USER(USER_ID)
        )
        ''')
```

```python
cur.execute('''
        CREATE TABLE GOALS(
            GOALID INT PRIMARY KEY,
            USER_ID INT,
            GOAL_NAME VARCHAR(100),
            START_DATE DATETIME,
            END_DATE DATETIME,
            GOAL_AMOUNT FLOAT,
            MONTHLY_SAVING_T FLOAT,
            GOAL_STATUS VARCHAR(50) CHECK(GOAL_STATUS IN("ACTIVE","PAUSED","ACHIEVED","EXPIRED","INACTIVE")),
            CREATED_AT DATETIME,
            UPDATED_AT DATETIME,
            FOREIGN KEY(USER_ID) REFERENCES USER(USER_ID))
            

     
        ''')
```

```python
def get_db():
    conn = sqlite3.connect("fintrackai.db")
    conn.row_factory = sqlite3.Row   # lets you access columns by name e.g. row["EMAIL"]
    return conn

# Inside every route:
conn = get_db()   # fresh connection, only for this request
cur = conn.cursor()
# ... do your queries ...
conn.close()      # always close when done
```
